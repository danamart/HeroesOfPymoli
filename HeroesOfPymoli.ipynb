## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
df = purchase_data.drop_duplicates(subset='SN', keep="last").reset_index(drop=True)

#establish bins for ages
bins = [0,9,14,19,24,29,34,39,100]
group_names = ["<10","10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorize the existing players using the age bins. Hint: use pd.cut()
df["Total Count"] = pd.cut(df['Age'], bins, labels=group_names)
#df["Total Count"].value_counts()

#Calculate the numbers and percentages by age group
Total_age = df["Age"].count()
less10 = df[df["Age"]< 10]["Purchase ID"].count()
ten14 = df[(df["Age"]>= 10) & (df["Age"]<= 14)]["Purchase ID"].count()
fifthteen19 = df[(df["Age"]>= 15) & (df["Age"]<= 19)]["Purchase ID"].count()
twenty24 = df[(df["Age"]>= 20) & (df["Age"]<= 24)]["Purchase ID"].count()
twentyfive29 = df[(df["Age"]>= 25) & (df["Age"]<= 29)]["Purchase ID"].count()
thirty34 = df[(df["Age"]>= 30) & (df["Age"]<= 34)]["Purchase ID"].count()
thirtyfive39 = df[(df["Age"]>= 35) & (df["Age"]<= 39)]["Purchase ID"].count()
more40 = df[df["Age"]>= 40]["Purchase ID"].count()

less10p = ((less10/Total_age)*100)
ten14p = ((ten14/Total_age)*100)
fifthteen19p = ((fifthteen19/Total_age)*100)
twenty24p = ((twenty24/Total_age)*100)
twentyfive29p = ((twentyfive29/Total_age)*100)
thirty34p = ((thirty34/Total_age)*100)
thirtyfive39p = ((thirtyfive39/Total_age)*100)
more40p = ((more40/Total_age)*100)

#Create a summary data frame to hold the results
age_demo = pd.DataFrame({"Labels" : ["<10","10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"],
                         "Total Count" : [less10, ten14, fifthteen19, twenty24, twentyfive29, thirty34, thirtyfive39,more40],
                        "Percentage of Players" : [less10p, ten14p, fifthteen19p, twenty24p, twentyfive29p, thirty34p, thirtyfive39p, more40p]})

#format the %
age_demo1 = age_demo.style.format({"Percentage of Players": "{:.2f}%"})
age_demo1
```

## Purchasing Analysis (Age)

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
#establish bins for ages
bins = [0,9,14,19,24,29,34,39,100]
group_names = ["<10","10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorize the existing players using the age bins. Hint: use pd.cut()
purchase_data["Purchase Count"] = pd.cut(purchase_data['Age'], bins, labels=group_names)

#calc purchase count
Total_agec = purchase_data["Age"].count()
less10c = purchase_data[purchase_data["Age"]< 10]["Price"].count()
ten14c = purchase_data[(purchase_data["Age"]>= 10) & (purchase_data["Age"]<= 14)]["Price"].count()
fifthteen19c = purchase_data[(purchase_data["Age"]>= 15) & (purchase_data["Age"]<= 19)]["Price"].count()
twenty24c = purchase_data[(purchase_data["Age"]>= 20) & (purchase_data["Age"]<= 24)]["Price"].count()
twentyfive29c = purchase_data[(purchase_data["Age"]>= 25) & (purchase_data["Age"]<= 29)]["Price"].count()
thirty34c = purchase_data[(purchase_data["Age"]>= 30) & (purchase_data["Age"]<= 34)]["Price"].count()
thirtyfive39c = purchase_data[(purchase_data["Age"]>= 35) & (purchase_data["Age"]<= 39)]["Price"].count()
more40c = purchase_data[purchase_data["Age"]>= 40]["Price"].count()

#avg PP
less10pp = purchase_data[purchase_data["Age"]< 10]["Price"].mean()
ten14pp = purchase_data[(purchase_data["Age"]>= 10) & (purchase_data["Age"]<= 14)]["Price"].mean()
fifthteen19pp = purchase_data[(purchase_data["Age"]>= 15) & (purchase_data["Age"]<= 19)]["Price"].mean()
twenty24pp = purchase_data[(purchase_data["Age"]>= 20) & (purchase_data["Age"]<= 24)]["Price"].mean()
twentyfive29pp = purchase_data[(purchase_data["Age"]>= 25) & (purchase_data["Age"]<= 29)]["Price"].mean()
thirty34pp = purchase_data[(purchase_data["Age"]>= 30) & (purchase_data["Age"]<= 34)]["Price"].mean()
thirtyfive39pp = purchase_data[(purchase_data["Age"]>= 35) & (purchase_data["Age"]<= 39)]["Price"].mean()
more40pp = purchase_data[purchase_data["Age"]>= 40]["Price"].mean()

#calc Total Purchase Value
less10pv = purchase_data[purchase_data["Age"]< 10]["Price"].sum()
ten14pv = purchase_data[(purchase_data["Age"]>= 10) & (purchase_data["Age"]<= 14)]["Price"].sum()
fifthteen19pv = purchase_data[(purchase_data["Age"]>= 15) & (purchase_data["Age"]<= 19)]["Price"].sum()
twenty24pv = purchase_data[(purchase_data["Age"]>= 20) & (purchase_data["Age"]<= 24)]["Price"].sum()
twentyfive29pv = purchase_data[(purchase_data["Age"]>= 25) & (purchase_data["Age"]<= 29)]["Price"].sum()
thirty34pv = purchase_data[(purchase_data["Age"]>= 30) & (purchase_data["Age"]<= 34)]["Price"].sum()
thirtyfive39pv = purchase_data[(purchase_data["Age"]>= 35) & (purchase_data["Age"]<= 39)]["Price"].sum()
more40pv = purchase_data[purchase_data["Age"]>= 40]["Price"].sum()

less10app = (less10pv/less10)
ten14app = (ten14pv/ten14)
fifthteen19app = (fifthteen19pv/fifthteen19)
twenty24app = (twenty24pv/twenty24)
twentyfive29app = (twentyfive29pv/twentyfive29)
thirty34app = (thirty34pv/thirty34)
thirtyfive39app = (thirtyfive39pv/thirtyfive39)
more40app = (more40pv/more40)
                       
#Create a summary data frame to hold the results
age_analysis = pd.DataFrame({"Labels" : ["<10","10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"],
                         "Purchase Count" : [less10c, ten14c, fifthteen19c, twenty24c, twentyfive29c, thirty34c, thirtyfive39c, more40c],
                            "Average Purchase Price": [less10pp, ten14pp, fifthteen19pp, twenty24pp, twentyfive29pp, thirty34pp, thirtyfive39pp, more40pp],
                            "Total Purchase Value": [less10pv, ten14pv, fifthteen19pv, twenty24pv, twentyfive29pv, thirty34pv, thirtyfive39pv, more40pv],
                            "Avg Total Purchase per Person" : [less10app, ten14app, fifthteen19app, twenty24app, twentyfive29app, thirty34app, thirtyfive39app, more40app]})
#format the $
age_analysis2 = age_analysis.style.format({"Average Purchase Price": "${:,.2f}",
                                               "Total Purchase Value": "${:,.2f}",
                                              "Avg Total Purchase per Person": "${:,.2f}"})
age_analysis2
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Run basic calculations to obtain the results in the table below
#group by SN
SN_gb = purchase_data.groupby('SN')
pc = SN_gb['Price'].count()
app = SN_gb['Price'].mean()
tpv = SN_gb['Price'].sum()

#Create a summary data frame to hold the results
#Sort the total purchase value column in descending order
top_spender = pd.DataFrame({"Purchase Count": pc, "Average Purchase Price": app, "Total Purchase Value": tpv})
top_spender2 = top_spender.sort_values(by = "Total Purchase Value", ascending = False)
top_spender2["Average Purchase Price"] = top_spender2["Average Purchase Price"].map("${:,.2f}".format)
top_spender2["Total Purchase Value"] = top_spender2["Total Purchase Value"].map("${:,.2f}".format)
top_spender2.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
popular_item = purchase_data.groupby(['Item ID', 'Item Name'])
pcp = popular_item['Price'].count()
ip = popular_item['Price'].mean()
tpvp = popular_item['Price'].sum()

#Create a summary data frame to hold the results
#Sort the purchase count column in descending order
popular_item_df = pd.DataFrame({"Purchase Count" : pcp, "Item Price": ip, "Total Purchase Value": tpvp})
popular_item_df2 = popular_item_df.sort_values(by = "Purchase Count", ascending = False)
popular_item_df2["Item Price"] = popular_item_df2["Item Price"].map("${:,.2f}".format)
popular_item_df2["Total Purchase Value"] = popular_item_df2["Total Purchase Value"].map("${:,.2f}".format)
popular_item_df2.head()
```
